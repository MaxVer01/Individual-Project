```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")
```

```python
netflix_df.head()
```

```python
# Calculate the minimum value in the column
min_value = netflix_df['release_year'].min()

# Calculate the maximum value in the column
max_value = netflix_df['release_year'].max()

print(min_value)
print(max_value)
```

```python
print(netflix_df.shape)
# Checking for the number of NaN values in each column
nan_count = netflix_df.isna().sum()

# Display the result
print(nan_count)
```

```python
#The missing values are not necessary to have in order to check for the duration of the movies so they will not be removed.
```

```python
#Since the data also consists out of TV-shows, we will filter on movies.
netflix_movies = netflix_df[netflix_df["type"] == "Movie"]
```

```python
netflix_movies.shape
```

```python
netflix_movies.head()
```

```python
# Filter for durations shorter than 60 minutes
short_movies = netflix_movies[netflix_movies.duration < 60]
short_movies.head()
```

```python
short_movies.shape
```

```python
# Define an empty list
colors = []
```

```python
# Iterating over rows of netflix_movies
for label, row in netflix_movies.iterrows() :
    if row["genre"] == "Children" :
        colors.append("red")
    elif row["genre"] == "Documentaries" :
        colors.append("blue")
    elif row["genre"] == "Stand-Up":
        colors.append("green")
    else:
        colors.append("black")
```

```python
colors[:10]
```

```python
# Set the figure style and initalize a new figure
fig = plt.figure(figsize=(12,8))

# Create a scatter plot of duration versus release_year
plt.scatter(netflix_movies.release_year, netflix_movies.duration, c=colors)

# Create a title and axis labels
plt.title("Movie Duration by Year of Release")
plt.xlabel("Release year")
plt.ylabel("Duration (min)")
```

```python
#From this point the guided project ends. Since im interested into looking into it more, and defining if there is a trend since 2010 we will look into it further.
```

```python
# Let's check all the unique genres present in the dataset
unique_genres = netflix_movies['genre'].unique()
unique_genres.sort()
unique_genres
```

```python
# Filter for movies from 2010 onwards
movies_since_2010 = netflix_movies[netflix_movies['release_year'] >= 2010]

# Calculate the average duration per year
average_duration_per_year = movies_since_2010.groupby('release_year')['duration'].mean().reset_index()

# Plotting
plt.figure(figsize=(12, 6))
plt.plot(average_duration_per_year['release_year'], average_duration_per_year['duration'], marker='o')
plt.title('Average Movie Duration on Netflix (2010 - Present)')
plt.xlabel('Year')
plt.ylabel('Average Duration (minutes)')
plt.grid(True)
plt.show()

```

```python
#Excluding speficic genres
# Excluding documentaries, children's movies, and stand-up from the dataset
excluded_genres = ['Documentaries', 'Children', 'Stand-Up']
movies_filtered = movies_since_2010[~movies_since_2010['genre'].isin(excluded_genres)]

# Calculate the average duration of movies for each year after excluding specified genres
average_duration_filtered = movies_filtered.groupby('release_year')['duration'].mean().reset_index()

average_duration_filtered.head(15)
```

```python
# Plotting
plt.figure(figsize=(12, 6))
plt.plot(average_duration_filtered['release_year'], average_duration_filtered['duration'], marker='o')
plt.title('Average Movie Duration on Netflix (2010 - Present)')
plt.xlabel('Year')
plt.ylabel('Average Duration (minutes)')
plt.grid(True)
plt.show()
```

```python
# we look into the Distribution of the Movie durations over the years
import seaborn as sns

# Plotting the distribution of movie durations
plt.figure(figsize=(12, 6))
sns.histplot(netflix_movies['duration'], kde=True, bins=30)
plt.title('Distribution of Netflix Movie Durations')
plt.xlabel('Duration (minutes)')
plt.ylabel('Frequency')
plt.show()
```

```python
recent_movies = netflix_movies[netflix_movies['release_year'] >= 2010]

plt.figure(figsize=(15, 8))
sns.boxplot(x='release_year', y='duration', data=recent_movies)
plt.xticks(rotation=90)  # Rotating the x-axis labels for better readability
plt.title('Distribution of Movie Durations by Release Year (2000 - Present)')
plt.xlabel('Release Year')
plt.ylabel('Duration (minutes)')
plt.show()
```

```python
# Filtering for TV Shows
tv_shows = netflix_df[netflix_df['type'] == 'TV Show']

# Filtering for years starting from 2000
tv_shows_2000 = tv_shows[tv_shows['release_year'] >= 2000]

# Counting the number of TV Shows per year
tv_shows_count = tv_shows_2000['release_year'].value_counts().sort_index()

# Plotting the graph
plt.figure(figsize=(10, 6))
tv_shows_count.plot(kind='bar', color='skyblue')
plt.title('Number of TV Shows Released Each Year (From 2000)')
plt.xlabel('Year')
plt.ylabel('Number of TV Shows')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
# Plotting the average duration of TV Shows per year
tv_shows_2010 = tv_shows[tv_shows['release_year'] >= 2010]

tv_shows_avg_duration = tv_shows_2010.groupby('release_year')['duration'].mean()

plt.subplot(1, 2, 2)
tv_shows_avg_duration.plot(kind='bar', color='lightgreen')
plt.title('Average Duration of TV Shows Each Year (From 2010)')
plt.xlabel('Year')
plt.ylabel('Average Duration (Seasons)')
plt.xticks(rotation=45)
plt.grid(axis='y')

plt.tight_layout()
plt.show()
```

```python
# Counting the number of movie releases per year since 2010
movies_since_2000 = netflix_movies[netflix_movies['release_year'] >= 2000]
movie_releases_count = movies_since_2000['release_year'].value_counts().sort_index()

# Plotting the number of movie releases each year since 2010
plt.figure(figsize=(10, 6))
movie_releases_count.plot(kind='bar', color='salmon')
plt.title('Number of Movie Releases Each Year (Since 2010)')
plt.xlabel('Year')
plt.ylabel('Number of Movie Releases')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
# Counting the number of short movies released each year
short_movies_count = short_movies['release_year'].value_counts().sort_index()

# Plotting the count of short movies released each year
plt.figure(figsize=(10, 6))
short_movies_count.plot(kind='bar', color='skyblue')
plt.title('Number of Short Movies Released Each Year')
plt.xlabel('Year')
plt.ylabel('Number of Short Movies')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
